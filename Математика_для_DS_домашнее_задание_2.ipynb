Дан объект в $2D$ пространстве

При помощи линейных отображений:
- Уменьшить объект в два раза,
- Повернуть на 130 градусов,
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)

```python
import numpy as np, sympy as sp, sklearn as skl, matplotlib.pyplot as plt
```

```python
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Уменьшить объект в два раза

вариант 1:

```python
B = np.array([[0.75, 0, 0], [0, 0.75, 0], [0, 0, 1]])
```

```python
tt = [];

for kk in A:
  tt.append(np.dot(B, kk))

tt
```

```python
tt = np.array(tt)

x = tt[:,0]
y = tt[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

вариант 2:

```python
res = np.dot(B, A.T)
res = res.T

res = np.array(res)

x = res[:,0]
y = res[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов

```python
import math

B2=[[math.cos(130*math.pi/180), math.sin(130*math.pi/180), 0],
 [-(math.sin(130*math.pi/180)), math.cos(130*math.pi/180), 0],
  [0, 0,1]]

B2
```

Вариант 1:

```python
ff = [];

for kk in tt:
  ff.append(np.dot(B2, kk))

ff
```

```python
ff = np.array(ff)

x = ff[:,0]
y = ff[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Вариан 2:

```python
res2 = np.dot(B2, tt.T)
res2 = res2.T

res2  = np.array(res2)

x = res2[:,0]
y = res2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]])

```python
v = [[0, 1, 0], [1, 0, 0], [0, 0, 1]]
v
```

```python
ss = [];

for kk in ff:
  ss.append(np.dot(v, kk))

ss
```

```python
ss = np.array(ss)

x = ss[:,0]
y = ss[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Найдите спектральное разложение матрицы:

```python
m = np.array([[1, 2],
              [2, 3]])

print(m)
```

```python
from sklearn.decomposition import PCA
```

```python
# собственное значение и собственный вектор
res = np.linalg.eig(m)
res
```

```python
# собственное значение
res[0]
```

```python
# собственный вектор
S = res[1]
S
```

```python
# диагональная матрица
D = np.diag(res[0])
D
```

```python
# тренспонированная ортогональная матрица собственных векторов
ST = np.transpose(S)
ST
```

```python
# проверка
A1 = np.dot(S, ST)
A1
```

```python
# спектральное разложение
A2 = np.dot(np.dot(S, D), ST)
A2
```

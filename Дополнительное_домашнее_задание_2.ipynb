Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.


```
### YOUR CODE HERE
```

```
import pandas as pd, numpy as np
```

```
# вычислим первую производную
```

```
from sympy import *
```

```
x = Symbol('x')
```

```
y = x**5 + 4*sin(2*x) + cos(3*x + 3)
```

```
yprime = y.diff(x)
yprime
```

```
yprime.subs(x, 1).evalf()
```

```
# можно еще и так
f = lambdify(x, yprime, 'numpy')
f(np.ones(1))
```

```
# вычислим вторую производную
```

```
yprime_2 = yprime.diff(x)
yprime_2
```

```
yprime_2.subs(x, 1).evalf()
```

```
# можно еще и так
f = lambdify(x, yprime_2, 'numpy')
f(np.ones(1))
```

Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```
### YOUR CODE HERE
```

```
# график первой производной
def f(x):
  return (np.sin(2*x + 1))**5

x = np.linspace(-5, 5, 1000)
dfdx = np.gradient(f(x), x)

fig, ax = plt.subplots()
ax.plot(x, f(x), label='f(x)')
ax.plot(x, dfdx, label="df/dx")
ax.legend()
plt.show()
```

```
# график второй производной
def f_2(x):
#  return dfdx
  return(np.gradient((np.sin(2*x + 1))**5))

x = np.linspace(-5, 5, 1000)
dfdx_2 = np.gradient(f_2(x), x)

fig, ax = plt.subplots()
ax.plot(x, f_2(x), label='f_2(x)')
ax.plot(x, dfdx_2, label="df/dx")
ax.legend()
plt.show()
```

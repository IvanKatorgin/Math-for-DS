Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:
- A1 - монеты 1 и 2 упали одной и той же стороной;
- A2 - монеты 2 и 3 упали одной и той же стороной;
- A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
# почему эти 3 события (A1, A2, A3) являются попарно независимыми
события (A1, A2, A3) являются попарно независимыми, если любые два из них независимы.
Из совместной независимости событий следует их попарная независимость:
P(A1∩A2)=P(A1)*P(A2)
P(A1∩A3)=P(A1)*P(A3)
P(A2∩A3)=P(A2)*P(A3)
```

```python
# вероятность, что какая-либо сторона любой монеты выпадет какой-либо стороной равна:
р = 1/2
```

т.к. вероятность выпадения какой-то стороны второй монеты не зависит от вероятности того, какой стороной выпадет первая монета, то вероятность события A1:

```python
p_A1(A1∩A2) = 1/2*1/2 = 1/4
```

Аналогично, вероятности событий А2 и А3 равны:

```python
#вероятность события А2:
p_A2(A2∩A3) = 1/2*1/2 = 1/4
```

```python
#вероятность события А2:
p_A3(A1∩A3) = 1/2*1/2 = 1/4
```

Таким образом мы доказали, что события (A1, A2, A3) являются попарно независимыми

```python
# почему эти 3 события (A1, A2, A3) не являются независимыми в совокупности
События называются независимыми в совокупности, если выполняется равенство:
P(A1∩A2∩A3)=P(A1)*P(A2)*P(A3)
Если это неравенство нарушается хотя бы для одного подмножества событий, то они зависимы в совокупности
```

```python
P(A1)*P(A2)*P(A3) = 1/2*1/2*1/2 = 1/8
p_A1(A1∩A2) = 1/4
p_A2(A2∩A3) = 1/4
p_A3(A1∩A3) = 1/4
# следовательно:
P(A1∩A2∩A3)=1/4!=P(A1)*P(A2)*P(A3)=1/8
1/4!=1/8
P(A1∩A2∩A3)!=P(A1)*P(A2)*P(A3)
```

Эти 3 события (A1, A2, A3) не являются независимыми в совокупности т.е. они зависимы в совокупности

Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распределения (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения. Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```python
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, t
```

Пять графиков плотностей для нормального распределения при разных параметрах

```python
# Разные параметры для нормального распределения
params = [(0, 0.7), (2, 1), (0, 2), (-2, 1), (0, 0.5)]

# Создание графиков
plt.figure(figsize=(12, 8))
for mu, sigma in params:
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    y = norm.pdf(x, mu, sigma)
    plt.plot(x, y, label=f'μ={mu}, σ={sigma}')

plt.legend()
plt.title('Графики плотности нормального распределения')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.show()
```

```python
normal = np.random.normal(loc=1, scale=10, size=1000)
```

```python
normal.mean(), normal.std(), normal.min(), normal.max()
```

Построим график распределения функции плотности и  распределения вероятности

```python
# зададим размер графика
plt.figure(figsize = (14,10))

# определим среднее значение и СКО
mean, std = normal.mean(), normal.std()

# зададим функции pdf и cdf на оси x
x = np.linspace(mean - 3 * std, mean + 3 * std, 1000)

# при этом возникает проблема: по оси y них разный масштаб
y1 = norm.pdf(x, mean, std)
y2 = norm.cdf(x, mean, std)

# эту проблему можно решить через функции subplots() и twinx()
# создадим сетку из одной ячейки
fig, ax_left = plt.subplots(nrows = 1, ncols = 1, figsize = (12,8))

# создадим новую ось с правой стороны
ax_right = ax_left.twinx()

# на оси x и левой оси y построим график функции плотности (pdf)
ax_left.plot(x, y1, label = 'P(x)')

# на оси x и правой оси y построим график функции распределения (cdf)
ax_right.plot(x, y2, color = 'orange', label = 'D(x)')

# также построим вертикальную прямую и точку
ax_right.vlines(x = mean, ymin = 0, ymax = 1, linewidth = 1, color = 'r', linestyles = '--')
ax_right.plot(mean, 0.5, marker = 'o', markersize = 5, markeredgecolor = 'r', markerfacecolor = 'r')

# из-за двух осей с легендой придется повозиться
fig.legend(loc = 'upper right',
           bbox_to_anchor = (1,1),
           bbox_transform = ax_right.transAxes,
           prop = {'size': 15})

plt.show()
```

Пять графиков плотностей для экспоненциального распределения при разных параметрах

```python
# Разные параметры для экспоненциального распределения
lambdas = [0.5, 1, 2, 5, 10]

# Создание графиков
plt.figure(figsize=(12, 8))
for lam in lambdas:
    x = np.linspace(0, 10, 100)
    y = expon.pdf(x, scale=1/lam)
    plt.plot(x, y, label=f'λ={lam}')

plt.legend()
plt.title('Графики плотности экспоненциального распределения')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.show()
```

Вариант 1

```python
def f(x, e, a, b):
    return np.e**a * np.sin(b*x) + x

params, covariancve = curve_fit(f, x, y)
err = np.sqrt(np.diag(covariancve))
print(params, err)

fit = []
for i in x:
  fit.append(f(i, *params))
fit = np.array(fit)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y, color='blue')
plt.plot(x, f(x, *params), color='red', ls='--')
plt.show()
```

```python
# Подбор функции как A*sin(a(x/b)+x)
def f(x, A, a, b):
    return A*np.sin(a*(x/b)+x)

params, covariancve = curve_fit(f, x, y)
err = np.sqrt(np.diag(covariancve))
print(params, err)

fit = []
for i in x:
  fit.append(f(i, *params))
fit = np.array(fit)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(x, y, color='blue')
plt.plot(x, f(x, *params), color='red', ls='--')
plt.show()
```
